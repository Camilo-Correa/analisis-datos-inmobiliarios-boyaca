```python
# Lista de palabras que quieres eliminar
palabras_a_eliminar = ['Duitama', 'Tunja', 'Sogamoso', 'Paipa']

# Función para eliminar palabras específicas de un texto
def limpiar_texto(texto):
    # Eliminamos cada palabra (ignorando mayúsculas y acentos)
    for palabra in palabras_a_eliminar:
        texto = re.sub(rf'\b{palabra}\b', '', texto, flags=re.IGNORECASE)
    # Limpiar espacios extra
    return re.sub(r'\s+', ' ', texto).strip()

# Aplicar la función a todas las columnas de texto
df_unificado['ubicacion_asociada'] = df_unificado['ubicacion_asociada'].apply(limpiar_texto)
df_unificado.tail()
```

```python
import numpy as np
import pandas as pd

# 1. Filtramos los inmuebles de arriendo en Duitama con precio faltante
filtros_arriendo = (
    (df_unificado1['Ciudad'].str.lower() == 'duitama') &
    (df_unificado1['finalidad'].str.lower() == 'arriendo') &
    (df_unificado1['precio'].isna())
)

# Guardamos esos índices
indices_a_imputar = df_unificado1[filtros_arriendo].index

# 2. Recorremos esos índices para imputar uno a uno
for idx in indices_a_imputar:
    fila = df_unificado1.loc[idx]
    tipo = fila['tipo_inmueble']
    estrato = fila['estrato']
    metros = fila['area m²']

    # 3. Buscamos inmuebles de VENTA en Duitama que coincidan en tipo y estrato
    venta_filtrada = df_unificado1[
        (df_unificado1['Ciudad'].str.lower() == 'duitama') &
        (df_unificado1['finalidad'].str.lower() == 'venta') &
        (df_unificado1['tipo_inmueble'] == tipo) &
        (df_unificado1['estrato'] == estrato) &
        (df_unificado1['precio'].notna()) &
        (df_unificado1['area m²'].notna())
    ]

    # 4. Si hay coincidencias, calculamos el valor m² promedio
    if not venta_filtrada.empty:
        valor_m2 = (venta_filtrada['precio'] / venta_filtrada['area m²']).mean()

        # 5. Calculamos el precio estimado de la propiedad
        precio_estimado = metros * valor_m2

        # 6. Tomamos el 0.8% como valor del arriendo mensual
        precio_arriendo = round(precio_estimado * 0.008)

        # 7. Asignamos el valor a la fila correspondiente
        df_unificado1.at[idx, 'precio'] = precio_arriendo
```

```python
#tratamiento de datos para antiguedad
import pandas as pd
import numpy as np
import re

def convertir_antiguedad_simple(valor):
    if pd.isna(valor):
        return np.nan
    valor = str(valor).lower()
    if 'menos de' in valor:
        return 0
    elif 'más de' in valor or 'mas de' in valor:
        numeros = re.findall(r'\d+', valor)
        return int(numeros[0]) if numeros else 30
    else:
        numeros = [int(n) for n in re.findall(r'\d+', valor)]
        return max(numeros) if numeros else np.nan

# Aplicar al DataFrame
df_unificado1['antigüedad'] = df_unificado1['antigüedad'].apply(convertir_antiguedad_simple)
```

```python
df_backup_3.head()
import pandas as pd
import re

# 1. Cargar tu dataframe original (df_nuevo3) y la API de barrios de Tunja
# df_nuevo3 = pd.read_csv("tu_archivo.csv")  # Ya lo tienes cargado

# 2. Descargar los barrios de Tunja desde el CSV de datos abiertos
url_tunja = "https://www.datos.gov.co/resource/4ryc-k7fx.csv"
df_tunja = pd.read_csv(url_tunja)

# 3. Limpiar texto para comparar las ubicaciones (para evitar problemas con mayúsculas/minúsculas y caracteres extraños)
def limpiar_texto(texto):
    if pd.isna(texto):
        return ""
    texto = texto.lower().strip()  # Convertir a minúsculas y quitar espacios innecesarios
    texto = re.sub(r'[^a-záéíóúüñ\s]', '', texto)  # Eliminar caracteres no deseados (acentos, puntuación)
    return texto

# 4. Crear un diccionario con los barrios de Tunja y sus coordenadas (latitud, longitud)
barrios_tunja_dict = {
    limpiar_texto(row['barrio']): {
        'barrio': row['barrio'],
        'latitud': row['latitud'],
        'longitud': row['longitud']
    }
    for _, row in df_tunja.iterrows()
}

# 5. Función para buscar un barrio en la API de Tunja y agregar latitud y longitud
def buscar_barrio_tunja(ubicacion):
    if pd.isna(ubicacion):
        return pd.Series([None, None, None])

    # Limpiar la ubicación de la columna
    ubicacion_limpia = limpiar_texto(ubicacion)

    # Buscar coincidencias con los barrios de Tunja en el diccionario
    for barrio_limpio_api, datos in barrios_tunja_dict.items():
        if barrio_limpio_api in ubicacion_limpia:
            # Si coincide, asignar barrio limpio, latitud y longitud
            return pd.Series([datos['barrio'], datos['latitud'], datos['longitud']])

    # Si no hay coincidencias, devolver NaN
    return pd.Series([None, None, None])

# 6. Aplicar la función solo a las filas de Tunja
def aplicar_lat_lon_tunja(row):
    # Si la ciudad es Tunja (lo verificamos con una simple búsqueda en la columna de ubicación)
    if "tunja" in limpiar_texto(row['ubicacion_asociada']):
        # Aplicamos la búsqueda y actualizamos barrio limpio, latitud y longitud
        return buscar_barrio_tunja(row['ubicacion_asociada'])
    else:
        # Si no es de Tunja, dejamos NaN para latitud y longitud, y no cambiamos barrio limpio
        return pd.Series([row['barrio_limpio'], None, None])

# 7. Aplicar la función al DataFrame y actualizar las columnas correspondientes
df_backup_3[['barrio_limpio', 'latitud', 'longitud']] = df_backup_3.apply(aplicar_lat_lon_tunja, axis=1)
```
